```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

df=pd.read_csv('protein_sequences.csv')

protein = ProteinAnalysis(df['sequence'].iloc[0])

aromaticity: protein.aromaticity()
instability: protein.instability_index()
charge: protein.charge_at_pH(7.0)

print(f"Molecular Weight: {protein.molecular_weight():.2f}")
print(f"Amino Acid Count: {protein.count_amino_acids()}")

```

```python
features = []

for seq in df['sequence']:
    protein = ProteinAnalysis(seq)
    features.append({
        'molecular_weight': protein.molecular_weight(),
        'gravy': protein.gravy(),
        'isoelectric_point': protein.isoelectric_point(),
        'helix_fraction': protein.secondary_structure_fraction()[0],
        'turn_fraction': protein.secondary_structure_fraction()[1],
        'sheet_fraction': protein.secondary_structure_fraction()[2],
    })

feature_df = pd.DataFrame(features)
feature_df.head()
```

```python
X=feature_df
Y=df['type']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=42)
modle= RandomForestClassifier()
modle.fit(X_train,Y_train)
prd=modle.predict(X_test)
acc=accuracy_score(Y_test,prd)
print(f"accurecy is :{acc*100}%")

```
