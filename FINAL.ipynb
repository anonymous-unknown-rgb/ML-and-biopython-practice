```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

df=pd.read_csv("bio_challenge_data.csv")

df.shape
df.head()

```

```python
from Bio.SeqUtils import gc_fraction
from Bio.SeqUtils.ProtParam import ProteinAnalysis

gc_scores = []
weights = []

for index, row in df.iterrows():
    # 1. Get the GC content number
    gc = gc_fraction(row['DNA_Sequence'])
    gc_scores.append(gc)
    
    
    clean_protein = row['Protein_Sequence'].replace("*", "")
    prot = ProteinAnalysis(clean_protein)
    weight = prot.molecular_weight()
    weights.append(weight)


df['GC_Content'] = gc_scores
df['Molecular_Weight'] = weights


df = df.drop(['Sample_ID', 'DNA_Sequence', 'Protein_Sequence'], axis=1)
```

```python
X=df.drop("Is_Pathogenic", axis=1)
Y=df["Is_Pathogenic"]

X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2)
modle=RandomForestClassifier()
modle.fit(X_train,Y_train)
pre=modle.predict(X_test)
accuracy=accuracy_score(Y_test,pre)
print(f"accuracy = {accuracy*100}%")
```
